**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
# Answer of Q2
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}. Current balance: {self.__balance}")
        else
            print("Invalid deposite amount!")

    def withdraw(self, amount):
        # check that at the time of withdrawal balance should not be less than zero:
        if amount > self.__balance:
            print("Error: Insufficient! Cannot withdraw, balance cannot be less than zero.")
        elif amount <=0:
            print("Invalid withdrawal amount!")
        else:
            self.__balance -= amount
            print(f"Withdrawn: {amount}. Remaining Balance: {self.__balance}")
```

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
class Book:
    def __init__(self, pages):
        self.pages = pages

    # Overloading the '+' operator
    def __add__(self, other):
        return self.pages + other.pages

# Testing:
b1 = Book(250)
b2 = Book(150)
total_pages = b1 + b2 
print(f"Total pages: {total_pages}")    #Output: 400
```

**Q15:** Kya Python default roop se Method Overloading (same method name with different arguments) support karta hai? Agar nahi, toh hum same name ke method ko multiple arguments ke sath kaise handle karte hain? Code likhein.

```python
# Python overloading ka traditional method ko support nahi karta jo directly aik hi function signature ko bar bar 
# define karta ha latest defined declaration simply overrides previous blocks. We simulate it using default argument
# rules or variable arguments (*args) 

class Computer:
    # Simulating overloading using optional default parameters
    def area(self, lenght, breadth=None):
        if breadth is not None:
            return lenght * breadth     # Rectangle area
            return lenght * lenght      # Square area

# Testing
calc = Computer()
print("Square Area:", calc.area(4))         # Output:  16
print("Rectangle Area", calc.area(4, 5))    # Output:  20
```

**Q17:** Q16 wali `Computer` class ko inherit karke do child classes `Laptop` aur `Desktop` banayein aur dono mein `process()` method ko mandatory implement karein.

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
from abc import ABC, abstractmethod

class Car(ABC):
    def start_engine(self): # concrete method
        print("Engine started... vroom!")

    @abstractmethod
    def drive(self):    #Abstract method
        pass

class ElectricCar(Car):
    def drive(self):
        print("Driving silently on battery power.")

# Testing:
tesla = ElectricCar()
tesla.start_engine()    # Triggers inherited concrete method
tesla.drive()           # Triggersimplented abstract method
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    def __init__(self, amount):
        self.amount = amount
    
    @abstractmethod
    def process_payment(self):
        pass

class CreditCardPayment(Payment):
    def __init__(self, amount, card_number):
        super().__init__(amount)
        self.card_number = card_number

    def process_payment(self):
        print(f"Processing credit card payment of ${self.amount} for Card Ending in {self.card_number[-4:]}.")

class UPIPayement(Payment):
    def __init__(self, amount, uupi_id):
        super().__init__(amount)
        self.upi_id = upi_id
```
